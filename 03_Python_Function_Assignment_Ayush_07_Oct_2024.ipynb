#### ***Q.1 . What is the difference between a function and a method in Python?***

Ans. :- **Function:** A function is a standalone block of code that performs a specific task. It can take inputs (parameters) and return an output (return value). Functions are defined using the *def* keyword outside of any class and they can be called independently, without needing to be associated with an object.
```
Example:-
def add(a, b):
    return a + b

result = add(3, 5)  # result is 8
```

**Method :-** A method is similar to a function but is associated with an object (an instance of a class). Methods are defined within a class and usually operate on the data contained within that class and they are called on objects and can access the instance (using self in instance methods) or the class (using cls in class methods).
```
Example:-
class Calculator:
    def add(self, a, b):
        return a + b
calc = Calculator()
result = calc.add(3, 5)  # result is 8
```

#### ***Q.2  Explain the concept of function arguments and parameters in Python.***

Ans. :- **Parameters :-** Parameters are the variables that are defined in a function's declaration. They act as placeholders for the values that will be passed to the function when it is called. They define what kind of input the function expects.
```
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")
```
**Arguments :-** Arguments are the actual values that you pass to a function when you call it. These values replace the parameters in the function. We can pass literals, variables, or expressions as arguments.
```
greet("Alice")  # "Alice" is the argument
```
***Types of Arguments***

**Positional Arguments:** These are passed in the same order as the parameters in the function definition.
```
def add(a, b):
    return a + b

result = add(3, 5)  # 3 and 5 are positional arguments
```
**Keyword Arguments:** These allow us to specify the parameter name, making the code clearer and allowing us to pass arguments in any order.
```
result = add(b=5, a=3)  # 5 is assigned to b, 3 to a
```
**Default Arguments:** These allow us to provide default values for parameters. If no argument is passed for that parameter, the default value is used.
```
def greet(name="World"):
    print(f"Hello, {name}!")

greet()         # Outputs: Hello, World!
greet("Alice")  # Outputs: Hello, Alice!
```
**Variable-length Arguments:** You can use *args and **kwargs to allow functions to accept an arbitrary number of positional or keyword arguments.
```
def sum_all(*args):
    return sum(args)

result = sum_all(1, 2, 3, 4)  # result is 10

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30)  # Outputs: name: Alice, age: 30
```

#### ***Q.4 What is the purpose of the `return` statement in a Python function?***

Ans :- The return statement in a Python function serves several important purposes:
1. **Exiting the Function :-** When the return statement is executed, the function immediately terminates, and control is returned to the caller. Any code after the return statement within that function will not be executed.
```
Example:
      def my_function():
          print("This will print.")
          return
          print("This will NOT print.")
my_function()  # Outputs: This will print.
```
2. **Returning a Value :-** The return statement allows a function to send back a value to the caller. This value can then be used or stored for further operations.
```
Example:
      def add(a, b):
          return a + b
      result = add(3, 5)  # result is 8
```
3. **Returning Multiple Values :-** Python functions can return multiple values as a tuple. When we use a comma-separated list of values, they are packed into a tuple and returned.
```
Example:
      def get_coordinates():
          return (10, 20)
      x, y = get_coordinates()  # x is 10, y is 20
```
4. **Optional Return :-** A function does not need to include a return statement. If no return is present, the function will return None by default when it reaches the end.
```
Example:
      def no_return():
          print("I don't return anything.")
      result = no_return()  # result is None
```

#### ***Q.5 What are iterators in Python and how do they differ from iterables?***

Ans :- iterables and iterators are fundamental to Python’s handling of data sequences, iterables are collections of data that can be looped over, whereas iterators are the objects that actually perform the iteration, yielding one item at a time and maintaining state.

**Iterables :-**
  * An iterable is any Python object that can be looped over (iterated) using a for loop or other iteration methods.
  * Examples of iterables include lists, tuples, strings, dictionaries, sets, and more.
  * An iterable implements the __iter__() method, which returns an iterator.
```
Example of an iterable:
      my_list = [1, 2, 3]
      for item in my_list:  # my_list is an iterable
          print(item)
```
**Iterators :-**
  * An iterator is an object that represents a stream of data; it generates values one at a time and maintains its state between each call.
  * An iterator implements two methods: __iter__() (which returns the iterator object itself) and __next__() (which returns the next value from the stream). When there are no more values to return, __next__() raises a StopIteration exception.
  * Unlike iterables, which can be reset and iterated over multiple times, an iterator can only be traversed once.
```
Example of an iterator:
      my_iterator = iter(my_list)  # Create an iterator from the list
      print(next(my_iterator))  # Outputs: 1
      print(next(my_iterator))  # Outputs: 2
      print(next(my_iterator))  # Outputs: 3
      # next(my_iterator) would now raise StopIteration
```
***Key Differences***
1.  **Definition:**
    * Iterable: An object that can be iterated over (e.g., lists, strings).
    * Iterator: An object that performs the iteration, producing values one at a time.
2.  **Methods:**
    * Iterable: Implements __iter__() method to return an iterator.
    * Iterator: Implements both __iter__() and __next__() methods.
3.  **State:**
    * Iterable: Can be iterated over multiple times.
    * Iterator: Can only be iterated once; after it has been exhausted, it cannot be reused.


#### ***Q.6 Explain the concept of generators in Python and how they are defined..***

Ans :- Generators in Python are a special type of iterable that allow us to create iterators in a more concise and memory-efficient way. They enable us to generate values on the fly without storing the entire sequence in memory, which is especially useful for handling large data sets or streams of data.

**Key Characteristics of Generators**
1.  **Lazy Evaluation:** Generators produce values one at a time and only when required. This means they do not compute all their values upfront, saving memory and computational resources.
2.  **State Retention:** Unlike regular functions that return a value and exit, generators maintain their state between successive calls, allowing them to resume where they left off.
3. **yield Keyword:** Generators are defined using the yield statement instead of return. When a generator function encounters yield, it produces a value and pauses its execution, saving its state for the next call.

**Defining a Generator**

  We define a generator similarly to a regular function but use the yield statement. Here’s a simple example:
```
Example of a Generator Function:
    def count_up_to(n):
        count = 1
        while count <= n:
            yield count  # Produce the current count and pause
            count += 1   # Increment the count
```
**Using a Generator**

  To use a generator, we can call it just like a regular function, but it returns a generator object. we can then iterate over this object using a for loop or the next() function.
```
Example of Using a Generator:
    counter = count_up_to(5)  # Create a generator
    for number in counter:  # Iterate over the generator
        print(number)
Output:
    1
    2
    3
    4
    5
```
**Generator Expressions**

In addition to defining generator functions, we can also create generators using generator expressions, which have a syntax similar to list comprehensions but use parentheses instead of square brackets.
```
Example of a Generator Expression:
      squares = (x * x for x in range(1, 6))
      for square in squares:
          print(square)
Output:
      1
      4
      9
      16
      25
```

#### ***Q.7 What are the advantages of using generators over regular functions?***

Ans :- Using generators offers several advantages over regular functions, particularly when it comes to managing memory and performance in certain scenarios.
1. **Memory Efficiency :-**

  Lazy Evaluation: Generators compute values on-the-fly and only produce one value at a time, which means they don’t require memory to store the entire sequence. This is particularly beneficial for working with large datasets or infinite sequences.

2. **Improved Performance :-**
  
  Reduced Overhead: Since generators yield values one at a time, we can start processing data immediately without waiting for the entire dataset to be computed. This can lead to faster performance in applications that can utilize the values as they are produced.

3. **State Retention :-**
    
  Maintaining State: Generators maintain their execution state between calls, which allows them to resume from where they left off. This contrasts with regular functions, which lose their state once they return a value.

4. **Cleaner and More Concise Code :-**

  Simplicity: Generators can simplify code by replacing complex iterator classes or functions with a more straightforward implementation. This can make our code easier to read and maintain.

5. **Pipelining :-**

  Chaining Generators: We can easily create a pipeline of generators that pass values through multiple stages of processing, making it easy to transform and filter data without creating intermediate lists.

6. **Support for Infinite Sequences :-**
  
  Generating Infinite Series: Generators can represent infinite sequences (e.g., Fibonacci numbers, prime numbers) without running out of memory, as they generate values indefinitely on demand.

7. **Better Handling of Large Data Streams :-**

  Processing Data in Chunks: Generators are ideal for reading and processing large files or streams of data in chunks, allowing for more responsive applications that can handle large inputs efficiently.

***Example of Using a Generator :-***
Consider a scenario where we need to process a large file line by line. Using a generator to read the file can be much more efficient than reading the entire file into memory at once.
```
  def read_large_file(file_name):
      with open(file_name) as file:
          for line in file:
              yield line.strip()  # Yield each line one at a time
  for line in read_large_file("large_file.txt"):
      process(line)  # Process each line without loading the entire file into memory
```


#### ***Q.8 What is a lambda function in Python and when is it typically used?***

Ans :- A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions defined with the def keyword, lambda functions are often used for simple operations that are defined in a single line. They can take any number of arguments but can only have one expression.
```
Syntax
    lambda arguments: expression
Example
add = lambda x, y: x + y
result = add(3, 5)  # result is 8
```
**Typical Uses of Lambda Functions**
  * **Short Functions:** Lambda functions are ideal for short, throwaway functions that you don't want to define with a full def statement.
  * **Higher-Order Functions:** They are commonly used in conjunction with higher-order functions like map(), filter(), and sorted(), where you need a small function to perform an operation on each element of a sequence.
```
Example with map():
      numbers = [1, 2, 3, 4]
      squares = list(map(lambda x: x ** 2, numbers))
      # Outputs: [1, 4, 9, 16]
Example with filter():
      even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
      # Outputs: [2, 4]
```
  * **Sorting:** Lambda functions can be used to define custom sorting logic in functions like sorted().
```
Example:
      points = [(1, 2), (4, 1), (2, 3)]
      sorted_points = sorted(points, key=lambda point: point[1])  # Sort by y-coordinate
```
  * **Callbacks:** They can be used as quick callback functions, especially in GUI applications or when defining event handlers.

**Advantages**
* Conciseness: They allow you to write quick functions without the overhead of a full function definition.
* Readability: For simple functions, they can enhance readability by reducing the amount of code.

**Limitations**
* Single Expression: Lambda functions can only contain a single expression, which limits their functionality compared to regular functions.
* No Documentation: They do not support documentation strings (docstrings), which can make them less self-documenting than standard functions.

#### ***Q.9 Explain the purpose and usage of the `map()` function in Python.***

Ans :- The map() function in Python is a built-in function that applies a given function to all items in an iterable (such as a list, tuple, or string) and returns a map object (which is an iterator). The purpose of map() is to simplify the process of applying a function to each element in an iterable without needing to write an explicit loop.

**Purpose**
  * Transformation: map() is commonly used to transform the items of an iterable by applying a function to each item, producing a new iterable with the transformed values.
  * Conciseness: It allows for cleaner and more concise code compared to using a loop for the same purpose.
```
Syntax
      map(function, iterable, ...)
```
* function: A function that defines the operation to be performed on each element.
* iterable: One or more iterable(s) to which the function is applied.

**Example Usage**
1.  Basic Example Here’s a simple example that squares each number in a list:
```
    numbers = [1, 2, 3, 4, 5]
    squared = map(lambda x: x ** 2, numbers)
    print(list(squared))  # Outputs: [1, 4, 9, 16, 25]
```
2.  Using a Defined Function You can also pass a regular function to map() instead of a lambda:
```
    def square(x):
        return x ** 2
    numbers = [1, 2, 3, 4, 5]
    squared = map(square, numbers)
    print(list(squared))  # Outputs: [1, 4, 9, 16, 25]
```
3.  Multiple Iterables map() can accept multiple iterables. The function must take as many arguments as there are iterables, and it applies the function in parallel to the items of each iterable.
```
    list1 = [1, 2, 3]
    list2 = [4, 5, 6]
    summed = map(lambda x, y: x + y, list1, list2)
    print(list(summed))  # Outputs: [5, 7, 9]
```
**When to Use map()**
  * When we need to apply a function to each item in an iterable and want to avoid explicit loops.
  * When we want to transform or process data in a clear and concise manner.
  * When working with functions that can be easily defined or expressed as lambda functions.

**Advantages**
  * Readability: Using map() can make your code more readable and expressive.
  * Efficiency: It can be more efficient than a loop, especially for large data sets, as it operates at the level of C speed in the background.

**Limitations**
  * Return Type: The result of map() is an iterator, so it must be converted to a list or another iterable type to see all values at once.
  * Less Flexible: For more complex transformations that involve multiple steps or require conditions, a list comprehension or a traditional loop might be more suitable.

#### ***Q.10  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?***

Ans :- In Python, map(), reduce(), and filter() are built-in functions that facilitate functional programming by applying functions to iterables. Each serves a distinct purpose, and understanding their differences can help you choose the right tool for a specific task.

1. map()
  * Purpose: Applies a function to all items in an iterable and returns a new iterable (map object) containing the results.
  * Usage: Best used when you want to transform each item in an iterable.
```    
Syntax:
        map(function, iterable, ...)
Example:
        numbers = [1, 2, 3, 4]
        squared = map(lambda x: x ** 2, numbers)
        print(list(squared))  # Outputs: [1, 4, 9, 16]
```
2. filter()
  * Purpose: Applies a function to all items in an iterable and returns a new iterable containing only the items for which the function returns True.
  * Usage: Best used when you want to filter items based on a condition.
```
Syntax:
        filter(function, iterable)
Example:
        numbers = [1, 2, 3, 4]
        even_numbers = filter(lambda x: x % 2 == 0, numbers)
        print(list(even_numbers))  # Outputs: [2, 4]
```
3. reduce()
  * Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.
  * Usage: Best used when you want to combine or aggregate values from an iterable into a single output.
```
Syntax:
      from functools import reduce
      reduce(function, iterable, [initializer])
Example:
      from functools import reduce
      numbers = [1, 2, 3, 4]
      product = reduce(lambda x, y: x * y, numbers)
      print(product)  # Outputs: 24 (1 * 2 * 3 * 4)
```
**Key Differences**
  * Return Type:
        map(): Returns an iterable with transformed items.
        filter(): Returns an iterable with selected items that meet a condition.
        reduce(): Returns a single value that is the result of combining items.
  * Functionality:
        map(): Transforms each item.
        filter(): Selects items based on a condition.
        reduce(): Aggregates items into a single result.
  * Input Requirements:
        map() and filter() take one iterable (or multiple for map()), while reduce() requires a single iterable and a function that takes two arguments.

#### ***Q.1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.***

Ans :-

````python
def sum_of_even_numbers(numbers):
  """
  This function takes a list of numbers as input and returns the sum of all even numbers in the list.
  """
  sum_even = 0
  for number in numbers:
    if number % 2 == 0:
      sum_even += number
  return sum_even

# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)
````

#### ***Q.4 Write a Python function that checks if a given number is prime or not from 1 to 200.***

````python
def is_prime(number):
  """
  This function checks if a given number is prime or not.
  """
  if number <= 1:
    return False
  for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
      return False
  return True

# Iterate through numbers from 1 to 200 and check for prime numbers
for num in range(1, 201):
  if is_prime(num):
    print(num, "is a prime number.")
````

#### ***Q.5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.***


````python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.num_terms:
            fib_num = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return fib_num
        else:
            raise StopIteration


# Example usage
num_terms = 10
fib_iterator = FibonacciIterator(num_terms)

for fib_num in fib_iterator:
    print(fib_num)
````

#### ***Q.6 Write a generator function in Python that yields the powers of 2 up to a given exponent.***


````python
def powers_of_2(exponent):
  """
  This generator function yields the powers of 2 up to a given exponent.
  """
  power = 0
  while power <= exponent:
    yield 2 ** power
    power += 1

# Example usage
exponent = 5
for num in powers_of_2(exponent):
  print(num)
````

#### ***Q.10  Create a Python program that uses `filter()` to remove all the vowels from a given string.***

````python
def remove_vowels(input_string):
    # Define a set of vowels
    vowels = 'aeiouAEIOU'

    # Use filter to remove vowels
    filtered_string = ''.join(filter(lambda char: char not in vowels, input_string))

    return filtered_string

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)  # Outputs: Hll, Wrld!
````
